```python
y_label=df['Diabetes_binary']
X_features=df.drop(['Diabetes_binary', 'AnyHealthcare'], axis=1)
#BMI ì´ìƒì¹˜ ì œê±°
dset=df['BMI']
q_25=np.percentile(dset.values, 25)
q_75=np.percentile(dset.values, 75)
iqr=(q_75-q_25)*1.5
smallest=q_25-iqr
largest=q_75+iqr
outlier_index=dset[(dset<smallest) | (dset>largest)].index
print('ì´ìƒì¹˜ ì¸ë±ìŠ¤ : ', outlier_index)
X_features.drop(outlier_index, axis=0, inplace=True)
y_label.drop(outlier_index, inplace=True)
```

```python
import xgboost as xgb
from xgboost import plot_importance
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

y_label=df['Diabetes_binary']
X_features=df.drop(['Diabetes_binary'], axis=1)

#X_features=StandardScaler().fit_transform(X_features)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit_transform(X_features)
X_features = scaler.fit_transform(X_features)
```

```python
#from imblearn.over_sampling import SMOTE

X_train, X_test, y_train, y_test=train_test_split(X_features, y_label, test_size=0.2, random_state=156)
print('ì›ë³¸ ë°ì´í„°:', X_train.shape, X_test.shape)

#smote=SMOTE(random_state=0)
#X_train, y_train=smote.fit_sample(X_train, y_train)
#print('SMOTE ì ìš©í•œ ë°ì´í„°:', X_train.shape, y_train.shape)
#print('SMOTE ì ìš©í•œ ë°ì´í„° ë¶„í¬:', pd.Series(y_train).value_counts())
```

```python
dtrain=xgb.DMatrix(data=X_train, label=y_train)
dtest=xgb.DMatrix(data=X_test, label=y_test)
```

```python
params={'max_depth':3,
       'eta':0.1,
       'objective':'binary:logistic',
       'eval_metric':'logloss',
       'early_stoppings':100
       }
num_rounds=400
```

```python
wlist=[(dtrain, 'train'), (dtest, 'eval')]
xgb_model=xgb.train(params=params, dtrain=dtrain, num_boost_round=num_rounds, early_stopping_rounds=100, evals=wlist)
```

```python
pred_probs=xgb_model.predict(dtest)
print('predict() ìˆ˜í–‰ ê²°ê³¼ê°’ì„ 10ê°œë§Œ í‘œì‹œ, ì˜ˆì¸¡ errorê°’ìœ¼ë¡œ í‘œì‹œë¨')
print(np.round(pred_probs[:10], 3))

preds=[1 if x> 0.5 else 0 for x in pred_probs]
print('ì˜ˆì¸¡ê°’ 10ê°œë§Œ í‘œì‹œ:', preds[:10])
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.metrics import f1_score, confusion_matrix, precision_recall_curve, roc_curve
import matplotlib.pyplot as plt
#163, 164íŽ˜ì´ì§€ eval_thresholdì½”ë“œ ë‚˜ì™€ìžˆìŒ ê·¸ê²ƒìœ¼ë¡œ ìˆ˜ì •
def eval_set_tpp(y_test, pred=None, pred_prob=None):
    confusion=confusion_matrix(y_test, pred)
    accuracy=accuracy_score(y_test, pred)
    precision=precision_score(y_test, pred)
    recall=recall_score(y_test, pred)
    f1=f1_score(y_test, pred)
    roc_auc=roc_auc_score(y_test, pred_prob)
    print('ì˜¤ì°¨í–‰ë ¬')
    print(confusion)
    print('ì •í™•ë„: {0:.4f}, ì •ë°€ë„: {1:4f}, ìž¬í˜„ìœ¨: {2:.4f},\
        F1: {3:4f}, AUC:{4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```

```python
def eval_set(y_test, pred=None):
    confusion=confusion_matrix(y_test, pred)
    accuracy=accuracy_score(y_test, pred)
    precision=precision_score(y_test, pred)
    recall=recall_score(y_test, pred)
    f1=f1_score(y_test, pred)
    print('ì˜¤ì°¨í–‰ë ¬')
    print(confusion)
    print('ì •í™•ë„: {0:.4f}, ìž¬í˜„ìœ¨: {2:.4f},\
        F1: {3:4f}'.format(accuracy, precision, recall, f1))
```

```python
from sklearn.preprocessing import Binarizer
thresholds=[0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65]

def get_eval_by_threshold(y_test, pred_prob, thresholds):
    for threshold_sp in thresholds:
        binarizer= Binarizer(threshold=threshold_sp).fit(pred_prob)
        pred_sp=binarizer.transform(pred_prob)
        print('threshold:', threshold_sp)
        eval_set(y_test, pred_sp)
        print("\n")
get_eval_by_threshold(y_test, pred_probs.reshape(-1,1), thresholds)
```

```python
 def roc_curve_plot(y_test, pred_probs):
    fig = plt.figure(figsize=(8,8))
    fpr, tpr, threshold = roc_curve(y_test, pred_probs)
    plt.plot(fpr, tpr, label='ROC')
    plt.plot([0,1], [0,1], 'k--', label='Random')
    start, end=plt.xlim()
    plt.xticks(np.round(np.arange(start, end, 0.1), 2))
    plt.xlim(0,1); plt.ylim(0,1)
    plt.xlabel('FPR(1-Sensitivity)');
    plt.ylabel('TPR(Recall)')
    plt.legend()
    fig.savefig('ROC_AUC_curve.png', dpi=fig.dpi)
```

```python
def precision_recall_curve_plot(y_test=None, pred_proba_c1=None):
        # threshold ndarrayì™€ ì´ thresholdì— ë”°ë¥¸ ì •ë°€ë„, ìž¬í˜„ìœ¨ ndarray ì¶”ì¶œ
        precisions, recalls, thresholds = precision_recall_curve( y_test, pred_proba_c1)
        # Xì¶•ì„ thresholdê°’ìœ¼ë¡œ, Yì¶•ì€ ì •ë°€ë„, ìž¬í˜„ìœ¨ ê°’ìœ¼ë¡œ ê°ê° Plot ìˆ˜í–‰. ì •ë°€ë„ëŠ” ì ì„ ìœ¼ë¡œ í‘œì‹œ
        plt.figure(figsize=(8,6))
        threshold_boundary = thresholds.shape[0]
        plt.plot(thresholds, precisions[0:threshold_boundary], linestyle='--', label='precision')
        plt.plot(thresholds, recalls[0:threshold_boundary],label='recall')
        # threshold ê°’ X ì¶•ì˜ Scaleì„ 0.1 ë‹¨ìœ„ë¡œ ë³€ê²½\n",
        start, end = plt.xlim()
        plt.xticks(np.round(np.arange(start, end, 0.1),2))
        # xì¶•, yì¶• labelê³¼ legend, ê·¸ë¦¬ê³  grid ì„¤ì •\n",
        plt.xlabel('Threshold value'); plt.ylabel('Precision and Recall value')
        plt.legend(); plt.grid()
        plt.show()
```
